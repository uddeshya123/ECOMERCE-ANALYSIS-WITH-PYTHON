```python
#import libraries
import pandas as pd 
import numpy as np 
import seaborn as sbn 
import matplotlib.pyplot as plt
```

```python
#import data set
df=pd.read_csv("ECOMERCE.csv",encoding="latin1")
```

```python
df.head()
```

```python
#getting information about data types in column
df.info()
```

```python
#check nan values
df.isna().sum()
```

```python
#check duplicates in data set
df.duplicated().sum()
```

```python
#change order date data type from str to date
df["Order Date"]=pd.to_datetime(df["Order Date"])
```

```python
#change ship date data type from str to date
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
```

```python
#again check info after change data type
df.info()
```

```python
df.head()
```

```python
#add 2 new columns month and year extract from order date
df["Order Month"]=df["Order Date"].dt.month
df["Order Year"]=df["Order Date"].dt.year
```

```python
df.head()
```

```python
#month vise sales
month_sales=df.groupby("Order Month")["Sales"].sum().reset_index()
month_sales
```

```python
#month vise sales in chart 
sbn.barplot(x="Order Month", y="Sales", data=month_sales)
plt.title("Sales by Month")
plt.show()
```

```python
type(month_sales)
```

```python
# sales by category
sales_catogery=df.groupby("Category")["Sales"].sum().reset_index()
sales_catogery
```

```python
#high sales by category
sbn.barplot(x="Category",y="Sales",data=sales_catogery)
plt.title("Sales by Category")
plt.show()
```

```python
#sales by sub-category
sales_sub_category=df.groupby("Sub-Category")["Sales"].sum()
sales_sub_category
```

```python
#top -5 highest sales in sub-category
sales_sub_category.sort_values(ascending=False).head()
```

```python
#month vise profit
month_profit=df.groupby("Order Month")["Profit"].sum()
month_profit
```

```python
#high profit in month
month_profit.sort_values(ascending=False).head(1)
```

```python
#low profit in month
month_profit.sort_values(ascending=False).tail(1)
```

```python
#profit by category
profit_by_category=df.groupby("Category")["Profit"].sum()
profit_by_category
```

```python
#highest profit in category
profit_by_category.sort_values(ascending=False).head(1)
```

```python
#lowest profit in category
profit_by_category.sort_values(ascending=False).tail(1)
```

```python
#profit by sub-category
profit_by_sub_category=df.groupby("Sub-Category")["Profit"].sum()
profit_by_sub_category
```

```python
#highest profit in sub-category
profit_by_sub_category.sort_values(ascending=False).head(1)
```

```python
#lowest profit in sub-category
profit_by_sub_category.sort_values(ascending=False).tail(1)
```

```python
df.head()
```

```python
#sales and profit by segment
sales_by_customer_segment=df.groupby("Segment")[["Sales","Profit"]].sum()
sales_by_customer_segment
```

```python
sales_by_customer_segment.plot(
    kind="bar",
    figsize=(8,5)
)
plt.title("Sales and Profit by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Amount")
plt.xticks(rotation=0)
plt.legend()
plt.show()
```

```python
#sales and profit ration over all
total_sales=df["Sales"].sum()
total_profit=df["Profit"].sum()
ratio=total_sales//total_profit
print("total sales",total_sales)
print("total profit",total_profit)
print("Sales and Profit ratio=",ratio)
```
